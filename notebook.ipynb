```python
from nogizaka import Blog
nanase = Blog('nanase.nishino')
nanase.crawl_urls()
```

```python
nanase.create_tables()
```

```python
nanase.add_all_details()
```

```python
cur = nanase.con.cursor()
```

```python
query = """
SELECT count(id)
FROM detail_contents
WHERE author='西野七瀬'
GROUP BY author
"""
cur.execute(query)
rows = cur.fetchall()
rows
```

```python
query = """
SELECT *
FROM(SELECT detail_id, url, file_path
    FROM images
    WHERE file_path='nanase.nishino.2e4180bebc39448ba4ec14103a4bf38a'
) a JOIN(SELECT id, url
    FROM detail_contents
) b ON a.detail_id=b.id
"""[1:-1]

cur.execute(query)
row = cur.fetchone()
for ro in row:
    print(ro)
```

```python
query = """
SELECT *
FROM (SELECT *
    FROM detail_contents
    WHERE author LIKE '西野七瀬'
)a JOIN(SELECT detail_id, url, file_path
    FROM images
)b ON a.id=b.detail_id
    ORDER BY char_length(content) asc
LIMIT 5
"""
cur.execute(query)
rows = cur.fetchall()
for row in rows:
    print(row[1])
    print(row[2])
    print(row[3])
    print(row[4])
    print(row[8])
```
